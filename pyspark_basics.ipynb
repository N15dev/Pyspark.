```python
data = [
    (101, "Asha", "Data Engineer", 80000),
    (102, "Ravi", "Analyst", 60000),
    (103, "Neha", "Data Engineer", 90000),
    (104, "Kiran", "Analyst", 55000)
]

columns = ["emp_id", "name", "role", "salary"]

df = spark.createDataFrame(data, columns)
df.printSchema()
```

```python
df.select("salary","emp_id")
df.select("salary","emp_id").show()
```

```python
df.filter(df.salary>70000)
df.filter(df.salary>70000).show()
```

```python
df.select("emp_id","salary")\
    .filter(df.salary>70000)\
    .show()
```
