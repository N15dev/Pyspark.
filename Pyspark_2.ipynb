```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

data=[(1,"Amit",60000),
      (2,"Neha",80000),
      (3,"Rahul",120000),
      (4,"Sara",750000)]

columns=["id","name","salary"]

df=spark.createDataFrame(data,columns)
df.show()
```

```python
from pyspark.sql.functions import col,when
df2=df.withColumn("salary_cateogry",
                    when(col("salary")>100000,"High")
                    .when(col("salary")>=60000,"Medium")
                    .otherwise("low")
                    )

df2.show()
```

```python
df2=df.withColumn("bonus",col("salary")*0.10)
df2.show()
```

```python
df3=df2.withColumn('total_salary',col('salary')+col("bonus"))
df3.show()
```

```python
df4=df3.withColumn("Performance Flag",
                   when(col("salary")>100000,"excellent")
                   .when(col("salary")>=60000,"good")
                   .otherwise("needs improvement")
                   )

df4.show()
```

```python
from pyspark.sql.functions import col, when

df5 = df.withColumn(
    "bonus_eligibility",
    
    when((col("salary") > 100000) & (col("name").startswith("R")), "Special Bonus")
    
    .when(col("salary") >= 100000, "Eligible")
    
    .otherwise("Partial")
)

df5.show()             
```
